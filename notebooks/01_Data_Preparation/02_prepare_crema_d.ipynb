```python

import os
import pandas as pd
```

```python
CREMA_PATH = r"D:\SER_Cross\data\crema_d"
```

```python
print(len(os.listdir(CREMA_PATH)))
print(os.listdir(CREMA_PATH)[:5])
```

```python
emotion_map = {
    "ANG": "angry",
    "HAP": "happy",
    "SAD": "sad",
    "NEU": "neutral",
    "FEA": "fear"
}
```

```python
records = []

for file in os.listdir(CREMA_PATH):
    if not file.lower().endswith(".wav"):
        continue

    parts = file.split("_")
    if len(parts) < 4:
        continue

    speaker_id = parts[0]
    emotion_code = parts[2]

    if emotion_code not in emotion_map:
        continue  # drop unwanted emotions

    records.append({
        "path": os.path.join(CREMA_PATH, file),
        "emotion": emotion_map[emotion_code],
        "speaker": f"crema_{speaker_id}",
        "dataset": "crema_d"
    })
```

```python
df_crema = pd.DataFrame(records)
df_crema.head()
```

```python
df_crema["emotion"].value_counts()
```

```python
output_path = r"D:\SER_Cross\data\processed\crema_d_metadata.csv"
df_crema.to_csv(output_path, index=False)

print("Saved:", output_path)
print("Total samples:", len(df_crema))
```
