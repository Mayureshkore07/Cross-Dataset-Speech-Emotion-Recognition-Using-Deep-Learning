```python
import os
import librosa
import soundfile as sf
from tqdm import tqdm
```

```python
INPUT_ROOT = r"D:\SER_Cross\data\crema_d"
OUTPUT_ROOT = r"D:\SER_Cross\data\processed\crema_d_audio"

os.makedirs(OUTPUT_ROOT, exist_ok=True)
```

```python
def process_audio(input_path, output_path, sr=16000):
    # load audio
    y, _ = librosa.load(input_path, sr=sr, mono=True)

    # trim silence
    y, _ = librosa.effects.trim(y, top_db=20)

    # normalize
    if y.max() > 0:
        y = y / abs(y).max()

    # save
    sf.write(output_path, y, sr)
```

```python
processed_count = 0

for file in tqdm(os.listdir(INPUT_ROOT)):
    if not file.lower().endswith(".wav"):
        continue

    input_wav = os.path.join(INPUT_ROOT, file)
    output_wav = os.path.join(OUTPUT_ROOT, file)

    process_audio(input_wav, output_wav)
    processed_count += 1

print("Processed CREMA-D files:", processed_count)
```
