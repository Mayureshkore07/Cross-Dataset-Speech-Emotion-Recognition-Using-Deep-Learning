```python
import os
import pandas as pd
```

```python
RAVDESS_PATH = r"D:\SER_Cross\data\ravdess"
```

```python
emotion_map = {
    "01": "neutral",
    "02": "neutral",   # calm → neutral
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fear"
}
```

```python
records = []

for folder in os.listdir(RAVDESS_PATH):
    folder_path = os.path.join(RAVDESS_PATH, folder)

    if not os.path.isdir(folder_path):
        continue

    if not folder.lower().startswith("actor_"):
        continue

    for file in os.listdir(folder_path):
        if not file.lower().endswith(".wav"):
            continue

        parts = file.split("-")
        if len(parts) < 3:
            continue

        emotion_code = parts[2]

        if emotion_code not in emotion_map:
            continue

        records.append({
            "path": os.path.join(folder_path, file),
            "emotion": emotion_map[emotion_code],
            "speaker": folder.lower(),
            "dataset": "ravdess"
        })
```

```python
df_ravdess = pd.DataFrame(records)
df_ravdess.head()
```

```python
df_ravdess = pd.DataFrame(records)
df_ravdess["emotion"].value_counts()
```

```python
output_path = r"D:\SER_Cross\data\processed\ravdess_metadata.csv"
df_ravdess.to_csv(output_path, index=False)

print("Saved:", output_path)
```
